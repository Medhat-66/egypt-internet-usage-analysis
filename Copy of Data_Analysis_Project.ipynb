```python
from faker import Faker
import pandas as pd
import numpy as np
import random

fake = Faker("ar_EG")
random.seed(42)
np.random.seed(42)

N = 6000
years = list(range(2015, 2025))

regions = [
    ("Cairo", 0.22, 0.95),
    ("Giza", 0.18, 0.90),
    ("Alexandria", 0.12, 0.92),
    ("Delta", 0.28, 0.60),
    ("Upper Egypt", 0.20, 0.40)
]

tariffs = {
    "WE": {
        "Fixed Broadband": [
            ("WE Start", 140, 10),
            ("WE Bronze", 250, 30),
            ("WE Silver", 400, 50),
            ("WE Gold", 700, 100)
        ],
        "Mobile Data": [
            ("WE Mobile 20GB", 120, 10),
            ("WE Mobile 40GB", 200, 20),
            ("WE Mobile 80GB", 350, 30)
        ]
    },
    "Vodafone": {
        "Fixed Broadband": [
            ("Vodafone Basic", 200, 30),
            ("Vodafone Plus", 350, 50),
            ("Vodafone Max", 650, 100)
        ],
        "Mobile Data": [
            ("Vodafone Red 20GB", 150, 15),
            ("Vodafone Red 40GB", 260, 25),
            ("Vodafone Red 80GB", 450, 40)
        ]
    },
    "Orange": {
        "Fixed Broadband": [
            ("Orange Home", 180, 30),
            ("Orange Extra", 330, 50),
            ("Orange Ultra", 600, 100)
        ],
        "Mobile Data": [
            ("Orange Go 20GB", 140, 15),
            ("Orange Go 40GB", 240, 25),
            ("Orange Go 80GB", 420, 40)
        ]
    },
    "Etisalat": {
        "Fixed Broadband": [
            ("Etisalat Home", 190, 30),
            ("Etisalat Power", 340, 50),
            ("Etisalat Extreme", 620, 100)
        ],
        "Mobile Data": [
            ("Etisalat Mix 20GB", 130, 15),
            ("Etisalat Mix 40GB", 230, 25),
            ("Etisalat Mix 80GB", 410, 40)
        ]
    }
}

data = []

for i in range(1, N + 1):

    age_group = np.random.choice(
        ["15-24", "25-34", "35-44", "45-54", "55+"],
        p=[0.22, 0.24, 0.20, 0.18, 0.16]
    )

    if age_group == "15-24":
        age = random.randint(15, 24)
        base_hours = random.uniform(4.5, 7.5)
        purpose = np.random.choice(["Social Media", "Streaming", "Gaming"], p=[0.45, 0.35, 0.20])
    elif age_group == "25-34":
        age = random.randint(25, 34)
        base_hours = random.uniform(4, 6.5)
        purpose = np.random.choice(["Work", "Streaming", "Social Media"], p=[0.40, 0.35, 0.25])
    elif age_group == "35-44":
        age = random.randint(35, 44)
        base_hours = random.uniform(3, 5.5)
        purpose = np.random.choice(["Work", "News", "Streaming"], p=[0.45, 0.30, 0.25])
    elif age_group == "45-54":
        age = random.randint(45, 54)
        base_hours = random.uniform(2.5, 4.5)
        purpose = np.random.choice(["News", "Communication"], p=[0.6, 0.4])
    else:
        age = random.randint(55, 75)
        base_hours = random.uniform(1.5, 3)
        purpose = np.random.choice(["News", "Communication"], p=[0.7, 0.3])

    region_name, _, urban = random.choices(regions, weights= [weg[1] for weg in regions])[0]

    is_internet_user = np.random.choice([1, 0], p=[0.83, 0.17])

    isp = np.random.choice(["WE", "Vodafone", "Orange", "Etisalat"],
                           p=[0.46, 0.22, 0.17, 0.15])

    connection = np.random.choice(["Mobile Data", "Fixed Broadband"], p=[0.72, 0.28])

    if is_internet_user:
        tariff_name, cost, speed = random.choice(tariffs[isp][connection])
        hours = base_hours + (speed / 100) * random.uniform(0.5, 1.2)
    else:
        tariff_name, cost, speed, hours, purpose = "None", 0, 0, 0, "None"

    data.append([
        i, i, age, random.choice(["Male", "Female"]),
        region_name, urban, is_internet_user,
        isp, connection, tariff_name,
        round(hours, 1), purpose,
        cost, speed, random.choice(years)
    ])

columns = [
    "record_id", "user_id", "age", "gender",
    "region_name", "urban_level", "is_internet_user",
    "isp", "connection_type", "tariff_name",
    "daily_hours", "usage_purpose",
    "monthly_cost_egp", "download_speed_mbps", "year"
]

df = pd.DataFrame(data, columns=columns)
df.to_csv("egypt_internet_usage.csv", index=False)
```

```python
# Get Data
import pandas as pd
import numpy as np

df_Data = pd.read_csv("/content/egypt_internet_usage.csv")
df_Data.head()
```

```python
df_Data.to_csv("Egypt_Internet_Usage.csv")
```

```python
df_Data['Internet_Speed_In_Region'] = df_Data['region_name'].map({"Alexandria":13.7, "Cairo": 15.8, "Upper Egypt" : 12, "Giza" : 17.1 , "Delta": 14})
```

```python
df_Data[['region_name', 'Internet_Speed_In_Region']].drop_duplicates()
```

```python
# 1. How has internet usage penetration changed over the years?
Internet_Users = df_Data[df_Data['is_internet_user'] == 1]
users = Internet_Users.groupby("year")['is_internet_user'].count().reset_index()
total = df_Data.groupby("year")['is_internet_user'].count().reset_index()
users['is_internet_user'] / total['is_internet_user']
```

```python
# 2. Which year experienced the highest growth in internet adoption?
Internet_Users = df_Data[df_Data['is_internet_user'] == 1]
users = Internet_Users.groupby("year")['is_internet_user'].count().reset_index(name="Count")
users['growth'] = None
for i in range(len(users)):
  if i > 0:
    users.loc[i , 'growth'] = users['Count'][i] - users['Count'][i-1]

users.loc[0, 'growth'] = -1
users.loc[users['growth'].max() == users['growth'], 'year']
```

```python
# 3.Has average daily internet usage increased over time?
Average_usage = df_Data.groupby('year')['daily_hours'].mean().reset_index()
Average_usage['increase Rate'] = None
for i in range(len(Average_usage)):
    if i > 0:
      Average_usage.loc[i , 'increase Rate'] = Average_usage['daily_hours'][i] - Average_usage['daily_hours'][i-1]
Average_usage
```

```python
# 4. Is internet usage growth slowing or accelerating in recent years?
Internet_Users = df_Data[df_Data['is_internet_user'] == 1]
users = Internet_Users.groupby("year")['is_internet_user'].count().reset_index()
total = df_Data.groupby("year")['is_internet_user'].count().reset_index(name ='Total Internet users')
rate = users.merge(total, on = 'year' ,how='left')
rate['growth'] = rate['is_internet_user'] / rate['Total Internet users']
rate['increase Rate'] = rate['growth'].diff()
rate
```

```python
#5. How does mobile vs fixed broadband adoption change over time?
Internet_Users_Only = df_Data[df_Data['is_internet_user']==1]
adoption = Internet_Users_Only.groupby(['connection_type', 'year'])['is_internet_user'].count().reset_index(name = 'users')
adoption['Rate'] = adoption.groupby('connection_type')['users'].diff()
adoption
```

```python
# 6. How does internet usage intensity vary by age group?
df_Data.groupby("Age_Group")['daily_hours'].mean().sort_index(ascending=True).reset_index(name= "Average Using")
```

```python
# 7.Which age group has the highest internet adoption rate?
Internet_Users_Only = df_Data[df_Data['is_internet_user']==1]
users = Internet_Users_Only.groupby('Age_Group')['is_internet_user'].count().reset_index(name = 'users')
total = df_Data.groupby("Age_Group")['is_internet_user'].count().reset_index(name ='Total Population')
adoption = users.merge(total, on='Age_Group')
adoption['Rate'] = adoption['users'] / adoption['Total Population']
adoption
```

```python
#8. Are younger users more likely to use mobile data than fixed broadband?
younger_users = df_Data[(df_Data['Age_Group'] =='15-24') & (df_Data['is_internet_user'] == 1)]
younger_users.groupby('connection_type')['is_internet_user'].count().sort_values(ascending=False)
```

```python
# 9. Which age groups are most likely to be non-internet users?
Non_Internet_Users_Only = df_Data[df_Data['is_internet_user']==0]
users = Non_Internet_Users_Only.groupby('Age_Group')['is_internet_user'].count().reset_index(name = 'users').sort_values(by = 'users', ascending=False)
users
```

```python
# 10. Which regions have the highest internet penetration? DESCENDING
Internet_Users = df_Data[df_Data['is_internet_user'] == 1]
total = df_Data.groupby('region_name')['is_internet_user'].count().reset_index(name ='Population')
users = Internet_Users.groupby("region_name")['is_internet_user'].count().reset_index(name ='Internet_users')
penetration = total.merge(users , on = 'region_name' )
penetration['penetration_Rate'] = penetration['Internet_users'] / penetration['Population']
penetration.sort_values(by = 'penetration_Rate', ascending=False)
```

```python
# 10. Which regions have the lowest internet penetration? ASCENDING
Internet_Users = df_Data[df_Data['is_internet_user'] == 1]
total = df_Data.groupby('region_name')['is_internet_user'].count().reset_index(name ='Population')
users = Internet_Users.groupby("region_name")['is_internet_user'].count().reset_index(name ='Internet_users')
penetration = total.merge(users , on = 'region_name' )
penetration['penetration_Rate'] = penetration['Internet_users'] / penetration['Population']
penetration.sort_values(by = 'penetration_Rate', ascending=True)
```

```python
#11. Is urbanization level correlated with internet usage?
internet_users = df_Data[df_Data['is_internet_user'] == 1]
internet_users['urban_level'].corr(other=internet_users['daily_hours'])
```

```python
#12. Do urban regions spend more on internet services than rural ones?
Regions = df_Data.groupby('region_name')['monthly_cost_egp'].mean().reset_index(name ='total_Money_Spent').sort_values(by='total_Money_Spent', ascending= False )
Regions
```

```python
#13. Which regions have the highest concentration of non-users?
Internet_Users = df_Data[df_Data['is_internet_user'] == 1]
total = df_Data.groupby('region_name')['is_internet_user'].count().reset_index(name ='Population')
users = Internet_Users.groupby("region_name")['is_internet_user'].count().reset_index(name ='Internet_users')
penetration = total.merge(users , on = 'region_name' )
penetration['penetration_Rate'] = penetration['Internet_users'] / penetration['Population']
penetration.sort_values(by = 'penetration_Rate', ascending=True)
```

```python
#14. What is the market share of each ISP?
with_ISP_Only = df_Data[(df_Data['isp'] != 'No is Sim Owner')]
total = with_ISP_Only.shape[0]
ISP = with_ISP_Only.groupby('isp')['user_id'].count().reset_index(name='users')
ISP['Market_Share'] = ISP['users'] / total
ISP
```

```python
#15. Does ISP preference differ by region?
Internet_Users = df_Data[df_Data['is_internet_user'] == 1]
users = Internet_Users.groupby(["region_name", 'isp'])['user_id'].count().reset_index(name ='total_subscriptors')
users
```

```python
#16. Which ISP has the highest average revenue per user (ARPU)?
ISP = df_Data.groupby('isp')['monthly_cost_egp'].mean().reset_index(name ='total_Money_Spent').sort_values(by='total_Money_Spent', ascending= False )
ISP
```

```python
#17.Are certain ISPs more popular among mobile or fixed broadband users?
ISP = df_Data.groupby(['isp', 'connection_type'])['user_id'].count().reset_index(name ='total_Users').sort_values(by='total_Users', ascending= False )
ISP
```

```python
#18.Does higher internet cost correspond to higher usage or speed?
df_Data['monthly_cost_egp'].corr(df_Data['daily_hours'])
```

```python
#19. What are the main reasons for not using the internet, and who are the non-users?
Non_users = df_Data[df_Data['Reason_Not_Using_Internet'] !='IS_Internet_User']
Non_users.groupby('Reason_Not_Using_Internet')['user_id'].count().reset_index(name='Number of non_user for this reason').sort_values(by = 'Number of non_user for this reason', ascending=False)
```

```python
#20. Which regions or age groups represent the highest potential for future internet growth?
Non_users = df_Data[df_Data['is_internet_user'] == 0]
Users = df_Data[df_Data['is_internet_user'] == 1]
Highest_Reasons_With_Age_users= Users.groupby(['region_name', 'Age_Group'])['user_id'].count().reset_index(name='user_count').sort_values(by = 'user_count', ascending=False)
Highest_Reasons_With_Age_Non_users= Non_users.groupby(['region_name', 'Age_Group'])['user_id'].count().reset_index(name='Non_users_count').sort_values(by = 'Non_users_count', ascending=False)
Potential_Growth = Highest_Reasons_With_Age_users.merge(Highest_Reasons_With_Age_Non_users , how='outer' , on=['region_name', 'Age_Group'])
Potential_Growth['Growth_Rate'] = Potential_Growth['Non_users_count'] / (Potential_Growth['user_count'] + Potential_Growth['Non_users_count'])
Potential_Growth
```
